```python
def fizz_buzz_encode(i):
    if i % 15 == 0:
        return 3
    elif i % 5 == 0:
        return 2
    elif i % 3 == 0:
        return 1
    else:
        return 0
    
def fizz_buzz_decode(i, pred):
    return [str(i), 'fizz', 'buzz', 'fizzbuzz'][pred]

def helper(i):
    print(fizz_buzz_decode(i, fizz_buzz_encode(i)))
    
for i in range(1, 16):
    helper(i)
```

```python
import numpy as np
import torch

NUM_DIGITS = 10

def binary_encode(i, num_digits):
    return np.array([i >> d & 1 for d in range(num_digits)])

trX = torch.tensor([binary_encode(i, NUM_DIGITS) for i in range(101, 2 ** NUM_DIGITS)]).float()
trY = torch.LongTensor([fizz_buzz_encode(i) for i in range(101, 2 ** NUM_DIGITS)])
```

```python
print(trX.shape)
print(trY.shape)
```

```python
trX[0, :]
```

```python
NUM_HIDDEN = 100
model = torch.nn.Sequential(
    torch.nn.Linear(NUM_DIGITS, NUM_HIDDEN),
    torch.nn.ReLU(),
    torch.nn.Linear(NUM_HIDDEN, 4)
)

loss_fn = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-2)

BATCH_SIZE = 128
for epoch in range(10000):
    for start in range(0, len(trX), BATCH_SIZE):
        end = start + BATCH_SIZE
        batchX = trX[start:end]
        batchY = trY[start:end]
        
        if torch.cuda.is_available():
            pass
        
        y_pred = model(batchX)
        loss = loss_fn(y_pred, batchY)
        
        print('Epoch: ', epoch, loss.item())
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```

```python
testX = torch.tensor([binary_encode(i, NUM_DIGITS) for i in range(1, 101)]).float()
with torch.no_grad():
    testY = model(testX)
    
predictions = zip(range(1, 101), list(testY.max(1)[1]))
```

```python
print([fizz_buzz_decode(i, x) for i, x in predictions])
```
